```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2,2,500)
y = np.linspace(-2,2,500)
xx, yy = np.meshgrid(x, y)
```

```python
def f(x, y):
    return x**2 - y**2 + 3 - 3* np.abs(x+y)**0.9
```

```python
z = np.vectorize(f)(xx,yy)
```

```python
plt.imshow(z)
plt.colorbar()
```
