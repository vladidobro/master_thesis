```python
import pymolzilla.processing as pm

import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

plt.rc('font', size=8) #controls default text size
plt.rc('axes', titlesize=8) #fontsize of the title
plt.rc('axes', labelsize=8) #fontsize of the x and y labels
plt.rc('xtick', labelsize=8) #fontsize of the x tick labels
plt.rc('ytick', labelsize=8) #fontsize of the y tick labels
plt.rc('legend', fontsize=8) #fontsize of the legend
```

```python
f1 = pm.cofe_room_t()
f2 = pm.cofe_low_t()
f1.df.drop(7, inplace=True)
f1.load()
f1.preprocess()
f1.process()
f2.load()
f2.preprocess()
f2.process()
```

```python
print('room T')
f1.analyze()
print('low T')
f2.analyze()
```

```python
ax = plt.gca()

ax.plot(f1.df['wavelength'], f1.df['Pp'])
ax.plot(f1.df['wavelength'], f1.df['Pm'])
ax.plot(f2.df['wavelength'], f2.df['Pp'])
ax.plot(f2.df['wavelength'], f2.df['Pm'])
```

```python
ax = plt.gca()

ax.plot(f1[5].data['phih'],f1[5].data[0.], 'o', label='0')
ax.plot(f1[5].data['phih'],f1[5].data[45.], 'o', label='45')

ax.plot(f1[5].data['phih'],f1[5].fit_anisotropy.pmldfit.results[0].fittedvalues, 'C0', label='fit')
ax.plot(f1[5].data['phih'],f1[5].fit_anisotropy.pmldfit.results[1].fittedvalues, 'C1', label='fit')
```

```python
def f(i,j):
    return (f1 if j==1 else f2).df['dPij'].apply(lambda x: x[i])

fig, (ax1, ax2) = plt.subplots(1,2)
for i in range(4):
    ax1.plot(f1.df['wavelength'], f(i,1))
    ax2.plot(f2.df['wa'])
```
